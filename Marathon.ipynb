```python
# Season for the data --> slower in summer than winter?
# Spring 3-5
# Summer 6-8
# fall 9-11
# winter 12-2

# Split two decimal for month
```

```python
df3.loc[:, 'race_month'] = df3['race_date'].str.split('.').str.get(1).astype(int)
```

```python
df3.loc[:, 'race_season'] = df3['race_month'].apply(
    lambda x: 'Spring' if x in range(3, 6)
    else 'Summer' if x in range(6, 9)
    else 'Fall' if x in range(9, 12)
    else 'Winter'
)
```
